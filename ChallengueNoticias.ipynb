##Se prepara el modelo (KNeighborsClassifier)

(caracteristicas = ['TABLAS A USAR']

)

```python
caracteristicas = ['Cantidad Verdadera','Afiliación Politica','Estado']

X = df[caracteristicas]

y = df['Etiqueta']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

modelo = KNeighborsClassifier(n_neighbors=5)

modelo.fit(X_train, y_train)

predit = modelo.predict(X_val)

accuracy = accuracy_score(y_val, predit)

print('Exactitud en la validación:', accuracy)
```
